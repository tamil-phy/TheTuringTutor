```python
def estimate_coef(x, y):
    # number of observations/points
    n = np.size(x)

    # mean of x and y vector
    m_x = np.mean(x)
    m_y = np.mean(y)

    # calculating cross-deviation and deviation about x
    SS_xy = np.sum(y*x) - n*m_y*m_x
    SS_xx = np.sum(x*x) - n*m_x*m_x

    # calculating regression coefficients
    w = SS_xy / SS_xx
    b = m_y - w*m_x

    return (w, b)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    'No_of_Butterflies': [1, 2, 3, 4, 5, 6, 7, 8],
    'Percent_of_Hunts': [30, 45, 51, 57, 60, 65, 70, 71]
}
```

```python
df = pd.DataFrame(data)
```

```python
X = df[['No_of_Butterflies']]
y = df['Percent_of_Hunts']
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
#plt.scatter(X_pred, y_pred, color='green', marker='x', label='Predictions')
plt.xlabel('Number of Butterflies')
plt.ylabel('Percent of Hunts')
plt.legend()
plt.title('Linear Regression for Prediction')
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
poly = PolynomialFeatures(degree=1)
X_poly = poly.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(X_poly,y)
```

```python
X_curve = np.linspace(min(X['No_of_Butterflies']), max(X['No_of_Butterflies']), 100).reshape(-1, 1)
X_curve_poly = poly.transform(X_curve)
y_curve = model.predict(X_curve_poly)

plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X_curve, y_curve, color='red', label='Polynomial Regression Curve')
#plt.scatter(X_pred, y_pred, color='green', marker='x', label='Predictions')
plt.xlabel('Number of Butterflies')
plt.ylabel('Percent of Hunts')
plt.legend()
plt.title('Polynomial Regression for Better Fit')
plt.show()
```

```python
intercept = model.intercept_
print(intercept)
```

```python
coefficients = model.coef_
intercept = model.intercept_
degree = len(coefficients) - 1
terms = [f"{coefficients[i]:.2f}x^{i}" for i in range(1, degree + 1)]
equation = f"y = {intercept:.2f} + " + " + ".join(terms)
```

```python
print(f"\nPolynomial Regression Equation: {equation}")
```
